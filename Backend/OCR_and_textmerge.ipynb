```python

def capture_frames_from_video(video_path, start_time, end_time, interval=10):
    cap = cv2.VideoCapture(video_path)
    frame_paths = []
    current_time = start_time
    while current_time <= end_time:
        cap.set(cv2.CAP_PROP_POS_MSEC, current_time * 1000)
        ret, frame = cap.read()
        if ret:
            # Create a unique filename for the frame
            frame_filename = f"/content/frame_{int(current_time)}s.png"
            frame_paths.append(frame_filename)

            # Save the frame as a PNG file in the Colab content folder
            cv2.imwrite(frame_filename, frame)

        current_time += interval
    cap.release()

    return frame_paths

def extract_text_from_frame(frame_path, previous_text=None, similarity_threshold=0.7):
    frame_image = cv2.imread(frame_path)
    ocr_text = pytesseract.image_to_string(frame_image)
    print(ocr_text)

    if previous_text:
        vectorizer = TfidfVectorizer().fit_transform([ocr_text, previous_text])
        similarity_matrix = cosine_similarity(vectorizer[0:1], vectorizer[1:2])
        similarity_score = similarity_matrix[0][0]

        if similarity_score > similarity_threshold:
            return None  # If similarity is above threshold, don't append the new text

    return ocr_text  # Return OCR text if it's sufficiently different
video_path = f"{cleaned_video_title}.mp4"
ocr_texts = []
all_frame_paths = []

for section in sections:
    start_time = section['start_time']
    end_time = section['end_time']

    captured_frame_paths = capture_frames_from_video(video_path, start_time, end_time)
    all_frame_paths.extend(captured_frame_paths)

    previous_text = None
    section_ocr_text = []

    for frame_path in captured_frame_paths:
        ocr_text = extract_text_from_frame(frame_path, previous_text)
        if ocr_text:
            section_ocr_text.append(ocr_text)
            previous_text = ocr_text  # Update the previous_text

    # Combine all OCR texts for the section
    ocr_texts.append(" ".join(section_ocr_text))
    print(f"Captured and processed frames for section from {start_time} to {end_time}")

# Add OCR text to DataFrame
df['ocr_text'] = ocr_texts
df['text'] = df['text'] + ' ' + df['ocr_text']
```

```python
df.to_csv(cleaned_video_title)
```
